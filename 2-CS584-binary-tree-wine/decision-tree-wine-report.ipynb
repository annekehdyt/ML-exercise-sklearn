# Decision Tree Classifier

Dataset : Wine Data Set
Source : Lichman, M. (2013). UCI Machine Learning Repository [http://archive.ics.uci.edu/ml]. Irvine, CA: University of California, School of Information and Computer Science.

Features of wine data are as follow :

    1) Alcohol 
    2) Malic acid 
    3) Ash 
    4) Alcalinity of ash 
    5) Magnesium 
    6) Total phenols 
    7) Flavanoids 
    8) Nonflavanoid phenols 
    9) Proanthocyanins 
    10)Color intensity 
    11)Hue 
    12)OD280/OD315 of diluted wines 
    13)Proline 
    
1. Read the data
2. Data preprocessing : Train_data, test_data [0.70, 0.30]
3. Decision Tree with Entropy
4. Decision Tree with Gini index
5. Model Evaluation using Test Set

## Read the data

```python
import pandas as pd
from pandas.tools.plotting import scatter_matrix
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn import metrics

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data"
names = ['wine', 
         'alcohol', 
         'malic-acid', 
         'ash', 
         'alca-ash', 
         'mg', 
         'tot-phenols', 
         'flavanoids', 
         'non-flava-phenols', 
         'proanthocyanins', 
         'col-intensity', 
         'hue', 
         'od-280-315', 
         'proline']
feature = ['alcohol', 
           'malic-acid', 
           'ash', 
           'alca-ash', 
           'mg', 
           'tot-phenols', 
           'flavanoids', 
           'non-flava-phenols', 
           'proanthocyanins', 
           'color-inten', 
           'hue', 
           'od-280-315', 
           'proline']
wine_df = pd.read_csv(url, 
                      names=names)
cls_name = ['Wine-1', 
            'Wine-2', 
            'Wine-3']

#wine_df.describe()
#wine_df.head(5)
```

## Preprocessing : Split train_data and test_data

```python
X_wine_train, X_wine_test, y_wine_train, y_wine_test = train_test_split(wine_df.iloc[:, 1:], 
                                                                        wine_df.iloc[:, 0], 
                                                                        test_size=0.30)
```

## DT (entropy)

```python
# Train data
# criterion = 'entropy'
# Max_depth = 1

DT_1_en = tree.DecisionTreeClassifier(criterion='entropy', max_depth=1)
DT_1_en.fit(X_wine_train, y_wine_train)

DT1en_dot_data = tree.export_graphviz(DT_1_en, 
                                        out_file=None, 
                                        feature_names=feature, 
                                        class_names=cls_name, 
                                        filled=True, rounded=True, 
                                        special_characters=True) 
DT1en_graph = graphviz.Source(DT1en_dot_data)

DT1en_graph
```

```python
# Test Data
y_pred_DT1en = DT_1_en.predict(X_wine_test)
acc_DT1en = accuracy_score(y_wine_test, y_pred_DT1en)
acc_norm_DT1en = accuracy_score(y_wine_test,y_pred_DT1en, normalize=False)
fpr, tpr, thresholds = metrics.roc_curve(y_wine_test, 
                                         y_pred_DT1en, 
                                         pos_label=2)
auc_DT1en = metrics.auc(fpr, tpr)

# Train Data
y_pred_tr_DT1en = DT_1_en.predict(X_wine_train)
acc_tr_DT1en = accuracy_score(y_wine_train, y_pred_tr_DT1en)
acc_norm_tr_DT1en = accuracy_score(y_wine_train,y_pred_tr_DT1en, normalize=False)
fpr, tpr, thresholds = metrics.roc_curve(y_wine_train, 
                                         y_pred_tr_DT1en, 
                                         pos_label=2)
auc_tr_DT1en = metrics.auc(fpr, tpr)
```

```python
# Train data
# criterion = 'entropy'
# Max_depth = 2

DT_2_en = tree.DecisionTreeClassifier(criterion='entropy', max_depth=2)
DT_2_en.fit(X_wine_train, y_wine_train)

# Test  Data
y_pred_DT2en = DT_2_en.predict(X_wine_test)
acc_DT2en = accuracy_score(y_wine_test, y_pred_DT2en)
acc_norm_DT2en = accuracy_score(y_wine_test,y_pred_DT2en, normalize=False)
fpr, tpr, thresholds = metrics.roc_curve(y_wine_test, 
                                         y_pred_DT2en, 
                                         pos_label=2)
auc_DT2en = metrics.auc(fpr, tpr)

# Train Data

y_pred_tr_DT2en = DT_2_en.predict(X_wine_train)
acc_tr_DT2en = accuracy_score(y_wine_train, y_pred_tr_DT2en)
acc_norm_tr_DT2en = accuracy_score(y_wine_train,y_pred_tr_DT2en, normalize=False)
fpr, tpr, thresholds = metrics.roc_curve(y_wine_train, 
                                         y_pred_tr_DT2en, 
                                         pos_label=2)
auc_tr_DT2en = metrics.auc(fpr, tpr)
```

# Feedback

<b>Criterion</b> {Gini Index, Entropy}
<br></br>
After several experiments, both Gini Index and Entropy has no significant difference when the tree depth more than 3 (since the total of class label is 3). However, for multi-class classification, the value of entropy could excess 1, thus, the use of Gini index is easier to look at. 

<b>Tree Depth</b>
<br></br>
There is not significance difference in the tree depth. However since there are 3 classes, the depth of the tree should at least 2 to give a higher accuracy.
<br></br>
Here, the max depth possible for wine datasets is 5. However, we could stop the depth from the depth of 3 or 4, since some of the leaves are already pure.

<b>Feature repetation</b>
<br></br>
The feature could appear more than once. In this dataset, the proline is re-appear on the same tree, with different threshold. Professor Bilgic once mentioned that this is possible when the value of features are not discrete value. 

<b>Random Feature Selection</b>
<br></br>
For several cases, the program would choose randomly among the similar entropy/gini index value. Thus, this could result in different evaluation result. The evidence that I had was the accuracy keeps changing everytime I re-execute the program. It could caused by the above reason.
