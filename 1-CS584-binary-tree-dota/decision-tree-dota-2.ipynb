# Decision Tree Classifier

Dataset : Dota2 Game result
Source : Lichman, M. (2013). UCI Machine Learning Repository [http://archive.ics.uci.edu/ml]. Irvine, CA: University of California, School of Information and Computer Science.

Each row of the dataset is a single game with the following features (in the order in the vector):
    1. Team won the game (1 or -1)
    2. Cluster ID (related to location)
    3. Game mode (eg All Pick)
    4. Game type (eg. Ranked)
    5 - end: Each element is an indicator for a hero. Value of 1 indicates that a player from team '1' played as that hero and '-1' for the other team. Hero can be selected by only one player each game. This means that each row has five '1' and five '-1' values.

```python
# Binary Tree Search

import pandas as pd
from pandas.tools.plotting import scatter_matrix
import numpy as np
import matplotlib.pyplot as plt
import graphviz
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.metrics import roc_auc_score
from sklearn import metrics

#pd.set_option('expand_frame_repr', 'default')

#Load data
dota_df_train = pd.read_csv('dota2Dataset/dota2Train.csv', header=None)
dota_df_test = pd.read_csv('dota2Dataset/dota2Test.csv', header=None)
```

```python
# Distribution
print(dota_df_train.groupby(0).size())
```

```python
# Assign the data by features and label

X_train = dota_df_train.iloc[:, 1:]
y_train = dota_df_train.iloc[:, 0]

X_test = dota_df_test.iloc[:, 1:]
y_test = dota_df_test.iloc[:, 0]
```

```python
# Max_depth = 1

DT1 = tree.DecisionTreeClassifier(max_depth=1)
DT1.fit(X_train, y_train)

DT1_data = tree.export_graphviz(DT1,out_file=None,  
                                        class_names=True, 
                                        filled=True, rounded=True, 
                                        special_characters=True) 
DT1_graph = graphviz.Source(DT1_data)
DT1_graph
```

```python
# Test Set
y_pred_DT1 = DT1.predict(X_test)
acc_DT1 = accuracy_score(y_test, y_pred_DT1)

# Train Set
y_pred_tr_DT1 = DT1.predict(X_train)
acc_tr_DT1 = accuracy_score(y_train, y_pred_tr_DT1)
```
